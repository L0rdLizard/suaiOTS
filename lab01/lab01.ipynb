```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f1 = 1180
f0 = 980
Vmod = 300
Vinf = 300
T = 1 / Vmod
A = 1

q = int(2 ** (Vinf / Vmod))

ti = np.linspace(0, T, 100)

e = ((A ** 2) * T) / 2
```

```python
def fi(i):
    if i == 0:
        return f0
    if i == 1:
        return f1
```

```python
def s(t, i):
    return A * np.cos(2 * np.pi * fi(i) * t)
```

```python
def integral(i):
    return np.trapz(s(ti, i) ** 2, ti)
```

```python
k = 0
fig, axs = plt.subplots(q, 1)
for i in fig.axes:
    i.grid()
    i.plot(ti, s(ti, k))
    k += 1
plt.show()

```

```python
x = np.linspace(0 , q - 1, q)
y1 = []
y2 = []
for i in x:
    y1.append(integral(i))
    y2.append(e)
```

```python
plt.plot(x, y1)
plt.plot(x, y2)
plt.legend(["Emod", "Eteor"])
plt.show()
```
