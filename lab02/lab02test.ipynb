```python
import matplotlib.pyplot as plt
import numpy as np
import random

f0 = 1650
f1 = 1950
Vmod = 300
Vinf = 300
T = 1 / Vmod
m = T * Vinf
q = 2 ** m
E = 1 / 600

W = 2 / T

f = np.arange(0, 10*W, 1)

S = lambda x, l: np.sqrt(2 * E / T) * (np.sinc((f - x) * T) + np.sinc((f + x) * T)) * np.exp(-1j * 2 * np.pi * l * f * T)
Smod = lambda x: np.sqrt(2 * E / T) * np.abs(np.sinc((f - x) * T) + np.sinc((f + x) * T))
#S0
plt.subplot(2, 2, 1)
plt.xlabel('S0')
plt.plot(f, np.abs(S(f0, 1)))
```

```python
#S1
plt.subplot(2, 2, 2)
plt.xlabel('S1')
plt.plot(f, np.abs(S(f1, 1)))
```
