```python
import matplotlib.pyplot as plt
import numpy as np
import math

f0 = 980
f1 = 1180

# f0 = 1650
# f1 = 1950
E = 1

Vmod = 300
Vinf = 300

T = 1 / Vmod

# множество сигналов
s1 = lambda t: np.sqrt(2*E/T) * np.cos(2 * np.pi * f0 * t)
s2 = lambda t: np.sqrt(2*E/T) * np.cos(2 * np.pi * f1 * t)

# множество ортонормированных функций
F1 = lambda t: np.sqrt(2/T) * np.cos(2 * np.pi * f0 * t)
F2 = lambda t: np.sqrt(2/T) * np.cos(2 * np.pi * f1 * t)

dt = (1 / f0) / 16
t = np.arange(0, T, dt)
#f = np.arange(0, 10, 0.01)

# res = 0
# 
# for i in t:
#     res = res + F1(i) * F2(i) * dt
# 
# res1 = np.sinc(2 * np.pi * (f0 * T + f1 * T)) + np.sinc(2 * np.pi * (f0 * T - f1 * T))

# print(res)
# print(res1)

# x1 = 0
# y1 = 0
# 
# x2 = 0
# y2 = 0
# 
# for i in t:
#    x1 = x1 + s1(i) * F1(i) * dt
#    y1 = y1 + s1(i) * F2(i) * dt
#    x2 = x2 + s2(i) * F1(i) * dt
#    y2 = y2 + s2(i) * F2(i) * dt

# координаты векторов коэффициентов для каждого сигнала
x1 = np.trapz(s1(t) * F1(t), t)
y1 = np.trapz(s1(t) * F2(t), t)
x2 = np.trapz(s2(t) * F1(t), t)
y2 = np.trapz(s2(t) * F2(t), t)

# print("s1(t) * F1(t) =", x1)
# print("s1(t) * F2(t) =", y1)
# print("s2(t) * F1(t) =", x2)
# print("s2(t) * F2(t) =", y2)

print("F1(t) * F1(t) =", np.trapz(F1(t) * F1(t), dx=dt))
print("F1(t) * F2(t) =", np.trapz(F1(t) * F2(t), dx=dt))
print("F2(t) * F1(t) =", np.trapz(F2(t) * F1(t), dx=dt))
print("F2(t) * F2(t) =", np.trapz(F2(t) * F2(t), dx=dt))
```

```python
k = (y1 - y2) / (x1 - x2)

perp = lambda x: -(1/k)*(x-(x1 + x2)/2)+(y1 + y2)/2

#print("x1= ", x1, " y1= ",y1," x2= ", x2," y2= ", y2," x3= ",(x1 + x2)/2, " y3= ", (y1 + y2)/2)

#pylab.subplot(1, 2, 1)
# comment 
# plt.plot([x1, math.ceil(x2), (x1 + x2)/2], [math.ceil(y1), y2, (y1 + y2)/2], 'go', [-1.5, 1.5, -1.5, 1.5], perp([-1.5, 1.5, -1.5, 1.5]))
# #plt.plot([x1, math.ceil(x2)], [math.ceil(y1), y2], 'go')
# 
# plt.axis([-2, 2, -2, 2])
# 
# ax = plt.gca()
# 
# ax.spines['left'].set_position('center')
# ax.spines['bottom'].set_position('center')
# ax.spines['top'].set_visible(False)
# ax.spines['right'].set_visible(False)
# 
# #plt.axis('equal')
# plt.show()
```

```python
x = F1(t)
y = F2(t)
S1 = s1(t)
S11x = np.trapz(S1 * x, dx=dt)
S11y = np.trapz(S1 * y, dx=dt)
plt.scatter(S11x, S11y)

x = F1(t)
y = F2(t)
S2 = s2(t)
S22x = np.trapz(S2 * x, dx=dt)
S22y = np.trapz(S2 * y, dx=dt)
plt.scatter(S22x, S22y)



size = 2

dx = 1 / size
areas = np.linspace(-1.0 - dx, 1.0 + dx, size + 1)

draw_lines = True

if draw_lines:
    for i, val in enumerate(areas):
        if i == 0 or i == size:
            continue
        plt.plot([val, val], [1.0 + dx, -1.0 - dx],
                 color="green", linestyle="--")
        plt.plot([1.0 + dx, -1.0 - dx], [val, val],
                 color="green", linestyle="--")

# plt.legend()
plt.plot([-1.5, 1.5, -1.5, 1.5], perp([-1.5, 1.5, -1.5, 1.5]))
plt.grid()
plt.show()
```
