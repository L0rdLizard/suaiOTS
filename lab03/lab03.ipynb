```python
import matplotlib.pyplot as plt
import numpy as np
import math

f0 = 980
f1 = 1180
E = 1

Vmod = 300
Vinf = 300

T = 1 / Vmod

# множество сигналов
s1 = lambda t: np.sqrt(2*E/T) * np.cos(2 * np.pi * f0 * t)
s2 = lambda t: np.sqrt(2*E/T) * np.cos(2 * np.pi * f1 * t)

# множество ортонормированных функций
F1 = lambda t: np.sqrt(2/T) * np.cos(2 * np.pi * f0 * t)
F2 = lambda t: np.sqrt(2/T) * np.cos(2 * np.pi * f1 * t)

dt = 1 / f0 / 100
t = np.arange(0, T, dt)
#f = np.arange(0, 10, 0.01)

res = 0

for i in t:
    res = res + F1(i) * F2(i) * dt

res1 = np.sinc(2 * np.pi * (f0 * T + f1 * T)) + np.sinc(2 * np.pi * (f0 * T - f1 * T))

print(res)
print(res1)

x1 = 0
y1 = 0

x2 = 0
y2 = 0

#for i in t:
#    x1 = x1 + s1(i) * F1(i) * dt
#    y1 = y1 + s1(i) * F2(i) * dt
#    x2 = x2 + s2(i) * F1(i) * dt
#    y2 = y2 + s2(i) * F2(i) * dt

# координаты векторов коэффициентов для каждого сигнала
x1 = np.trapz(s1(t) * F1(t), t, dt)
y1 = np.trapz(s1(t) * F2(t), t, dt)
x2 = np.trapz(s2(t) * F1(t), t, dt)
y2 = np.trapz(s2(t) * F2(t), t, dt)

print("s1(t) * F1(t) =", x1)
print("s1(t) * F2(t) =", y1)
print("s2(t) * F1(t) =", x2)
print("s2(t) * F2(t) =", y2)
```

```python
k = (y1 - y2) / (x1 - x2)

perp = lambda x: -(1/k)*(x-(x1 + x2)/2)+(y1 + y2)/2

#print("x1= ", x1, " y1= ",y1," x2= ", x2," y2= ", y2," x3= ",(x1 + x2)/2, " y3= ", (y1 + y2)/2)

#pylab.subplot(1, 2, 1)
plt.plot([x1, math.ceil(x2), (x1 + x2)/2], [math.ceil(y1), y2, (y1 + y2)/2], 'go', [-1.5, 1.5, -1.5, 1.5], perp([-1.5, 1.5, -1.5, 1.5]))
#plt.plot([x1, math.ceil(x2)], [math.ceil(y1), y2], 'go')

plt.axis([-2, 2, -2, 2])

ax = plt.gca()

ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

#plt.axis('equal')
plt.show()
```
